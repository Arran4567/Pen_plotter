```python
import os
import skimage 
import numpy as np
import matplotlib.pyplot as plt
from pygcode import *
from skimage import measure
from skimage import io
from skimage import data
from skimage.color import rgb2gray
```

```python
def import_image(path):
    file = os.path.join(path)
    img = io.imread(file)
    io.imshow(img) 
    io.show() 
    return img
```

```python
image = import_image("smile.png")
```

```python
def img_to_gray(original):
    grayscale = rgb2gray(original)
    fig, axes = plt.subplots(1, 1, figsize=(4, 4))

    axes.imshow(grayscale, cmap=plt.cm.gray)
    axes.set_title("Grayscale")

    fig.tight_layout()
    plt.show()
    return grayscale
```

```python
gray_img = img_to_gray(image)
```

```python
def find_contours(image):
    img = np.flip(image)
    
    # Find contours at a constant value of 0.9
    contours = measure.find_contours(img, .9)
    
    # Display the image and plot all contours found
    fig, ax = plt.subplots()

    for n, contour in enumerate(contours):
        ax.plot(contour[:, 1], contour[:, 0], linewidth=1)

    ax.axis('image')
    plt.show()
    return contours
```

```python
contours = find_contours(gray_img)
len(contours)
```

```python
def get_GCode(contours):
    i = 0
    to_coords = lambda coords: {'X': coords[0], 'Y': coords[1]}
    for contour in contours:
        for coords in contour:
            if(i == 0):
                print("%s" % GCodeRapidMove(**to_coords(coords)))
                i=1
            else:
                print("%s" % GCodeLinearMove(**to_coords(coords)))
        i=0
        print("-" * 40)
    return ""
```

```python
get_GCode(contours)
```
