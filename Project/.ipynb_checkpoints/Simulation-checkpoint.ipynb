```python
import turtle
import re
from pygcode import *
```

```python
def setup_turtle():
    t = turtle.Turtle()
    
    s = t.getscreen()
    s.screensize()
    s.setup(width = 1.0, height = 1.0)
    canvas = s.getcanvas()
    root = canvas.winfo_toplevel()
    root.overrideredirect(1)
    
    t.hideturtle()
    t.penup()
    t.shape("classic")
    t.getscreen()._root.attributes('-topmost', 1)
    t.showturtle()
    t.speed(10)
    return t,s
```

```python
t, s = setup_turtle()
```

```python
def set_home(x, y):
    t.setposition(x, y)
    return x, y
```

```python
x_home, y_home = set_home(-955, -535)
```

```python
def move_pos(x, y):
    t.penup()
    t.setposition(x_home + x, y_home + y)
```

```python
def draw_line(x, y):
    t.pendown()
    t.setposition(x_home + x, y_home + y)
```

```python
def convert_gcode(gcode, scale, x_fix, y_fix):
    with open(gcode, 'r') as fh:
        for line_text in fh.readlines():
            line = Line(line_text)
            
            print(line)  # will print the line (with cosmetic changes)
            
            x_substring = int(re.search("X(.*?) ", str(line)).group(1))
            y_substring = int(re.search("Y(.*?) ", str(line)).group(1))
            
            print(x_substring)
            print(y_substring)
                
            if str(line)[0:3] == "G00":
                print("G00")
                move_pos(x_substring, y_substring)
            elif str(line)[0:3] == "G01":
                print("G01")
                draw_line(x_substring, y_substring)
            elif str(line)[0:3] == "G03":
                print("G03")
            else:
                print("Please enter valid instructions.")
            
            print('-'*40)  # will print the line (with cosmetic changes)
```

```python
convert_gcode("output.gcode", 300, 2000, 1200)
```

```python
def draw_arc(turtle, x, y, r, start, end, scale, x_fix, y_fix):
    x_real = x*scale-x_fix
    y_real = y*scale-y_fix
    r_real = r*scale
    extent = end-start
    print(x_real, y_real)
    
    turtle.penup()
    turtle.setposition(x_real, y_real)
    turtle.pendown()
    turtle.setheading(start)
    turtle.circle(r_real, extent)
```

```python
def draw_line(turtle, x, y, scale, x_fix, y_fix):
    x_real = x*scale-x_fix
    y_real = y*scale-y_fix
    print(x_real, y_real)
    turtle.pendown()
    turtle.setposition(x_real, y_real)
    turtle.penup()
```

```python
def clear_screen(s):
    s.reset()
    t.hideturtle()
    t.penup()
    t.setposition(-955, -535)
```

```python
def draw_smiley():
    t.showturtle()
    t.speed(1)
    t.up()
    t.goto(0, -100)  # center circle around origin
    t.down()
    
    t.begin_fill()
    t.fillcolor("yellow")  # draw head
    t.circle(100)
    t.end_fill()

    t.up()
    t.goto(-67, -40)
    t.setheading(-60)
    t.width(5)
    t.down()
    t.circle(80, 120)  # draw smile

    t.fillcolor("black")

    for i in range(-35, 105, 70):
        t.up()
        t.goto(i, 35)
        t.setheading(0)
        t.down()
        t.begin_fill()
        t.circle(10)  # draw eyes
        t.end_fill()
    
    t.up()
    t.goto(-67, -40)
```

```python
def close_sim(s):
    s.exitonclick()
    convert_gcode("output.gcode", 300, 2000, 1200)
```
